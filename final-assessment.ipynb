```python
import pandas as pd
```

```python
df = pd.read_csv('FRMGHAM2_ASSESSMENT_TRAIN.csv')
```

```python
labels=['Underweight','Normal Weight','Over Weight','Obesity Class 1','Obesity Class 2','Obesity Class 3']
cut_bins = [0, 18.49, 24.9, 29.9, 34.9,39.9,60]
df['BMI Bin'] = pd.cut(df['BMI'], bins=cut_bins, labels=labels)

```

```python
df['BMI Bin'].value_counts()
```

```python
labels=['< 29','30-39','40-49','> 50']
cut_bins = [0, 29, 39, 49, 90]
df['Age Bin'] = pd.cut(df['AGE'], bins=cut_bins,labels=labels)

```

```python
df['Age Bin'].value_counts()
```

```python
labels=['TOTCHOL < 200','TOTCHOL > 200']
cut_bins = [0, 200, 800]
df['TOTCHOL Bin'] = pd.cut(df['TOTCHOL'], bins=cut_bins, labels=labels)
```

```python
labels=['LDLC < 130','LDLC > 130']
cut_bins = [0, 130, 700]
df['LDLC Bin'] = pd.cut(df['LDLC'], bins=cut_bins, labels=labels)
```

```python
labels=['HDLC < 40','HDLC > 40']
cut_bins = [0, 40, 300]
df['HDLC Bin'] = pd.cut(df['HDLC'], bins=cut_bins, labels=labels)
```

```python
labels=['SYSBP < 120','SYSBP > 120']
cut_bins = [0, 120, 400]
df['SYSBP Bin'] = pd.cut(df['SYSBP'], bins=cut_bins, labels=labels)
```

```python
labels=['DIABP < 80','DIABP > 80']
cut_bins = [0, 80, 300]
df['DIABP Bin'] = pd.cut(df['DIABP'], bins=cut_bins, labels=labels)
```

```python
df.head(20)
```

```python
df_ar = df[df['ANYCHD']=='Yes'].copy()
#df_ar = df_ar[["RANDID", "DIABETES", "CURSMOKE", "BPMEDS", "BMI Bin", "Age Bin" ,"ANYCHD"]]
df_ar = df_ar[["RANDID", "DIABETES", "CURSMOKE", "BPMEDS"]]
```

```python
df_ar.shape
```

```python
df_ar['AGE'] = df['Age Bin'].copy()
df_ar['BMI'] = df['BMI Bin'].copy()
df_ar['TOTCHOL'] = df['TOTCHOL Bin'].copy()
df_ar['HDLC'] = df['HDLC Bin'].copy()
df_ar['LDLC'] = df['LDLC Bin'].copy()
df_ar['SYSBP'] = df['SYSBP Bin'].copy()
df_ar['DIABP'] = df['DIABP Bin'].copy()
```

```python
df_ar
```

```python
df_trans = pd.DataFrame()

for index, data in df_ar.iterrows():
   if data["DIABETES"] == "Yes":
     df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["DIABETES"] })])
   if data["CURSMOKE"] == "Yes":
     df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["CURSMOKE"] })])
   if data["BPMEDS"] == 1:
     df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["BPMEDS"] })])
   #if data["ANYCHD"] == "Yes":
     #df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["ANYCHD"] })])
    
   if data["AGE"] == "< 29":
     df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["< 29"] })])
   if data["AGE"] == "30-39":
     df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["30-39"] })])
   if data["AGE"] == "40-49":
     df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["40-49"] })])
   if data["AGE"] == "> 50":
     df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["> 50"] })])
    
   if data["BMI"] == "Underweight":
     df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["Underweight"] })])
   if data["BMI"] == "Normal Weight":
     df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["Normal Weight"] })])
   if data["BMI"] == "Over Weight":
     df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["Over Weight"] })])     
   if data["BMI"] == "Obesity Class 1":
     df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["Obesity Class 1"] })])
   if data["BMI"] == "Obesity Class 2":
     df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["Obesity Class 2"] })])
   if data["BMI"] == "Obesity Class 3":
     df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["Obesity Class 3"] })])
    
   if data["TOTCHOL"] == "TOTCHOL < 200":
      df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["TOTCHOL < 200"] })])
   if data["TOTCHOL"] == "TOTCHOL > 200":
      df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["TOTCHOL > 200"] })])
    
   if data["LDLC"] == "LDLC < 130":
      df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["LDLC < 130"] })])
   if data["LDLC"] == "LDLC > 130":
     df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["LDLC > 130"] })])
        
   if data["HDLC"] == "HDLC < 40":
      df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["HDLC < 40"] })])
   if data["HDLC"] == "HDLC > 40":
     df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["HDLC > 40"] })])
        
   if data["SYSBP"] == "SYSBP < 120":
      df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["SYSBP < 120"] })])
   if data["SYSBP"] == "SYSBP > 120":
     df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["SYSBP > 120"] })]) 
        
   if data["DIABP"] == "DIABP < 80":
      df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["DIABP < 80"] })])
   if data["DIABP"] == "DIABP > 80":
     df_trans = pd.concat([df_trans, pd.DataFrame({ "RANDID": [data["RANDID"]], "Condition": ["DIABP > 80"] })]) 
df_trans = df_trans.reset_index(drop=True)
```
